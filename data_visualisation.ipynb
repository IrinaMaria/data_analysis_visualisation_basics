## Loading the equirements

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt

%matplotlib inline  
```

## Connecting to the dataset 

```python
df = pd.read_csv ("data/life_expectancy_and_income.csv")
```

 ## Built-in-plotting - pandas will use matplotlib

### Histogram

An example of a histogram using the `income` data. This is using the `plot()` method on a *column*. We pass in an argument to specify the `kind` of plot.

```python
df ["income_per_person"].plot (kind = "hist", edgecolor = "white", bins = 40 )      # plot income per person
```

### Box Plot 

```python
df ["income_per_person"].plot (kind = "box")      
```

### Scatter Plot 

Using the `plot()` method on a whole data frame to generate a scatterplot.

```python
df.plot(kind = "scatter", x = "income_per_person", y = "life_expectancy")
```

### Bar chart 

```python
last_20 = df.loc[df["year"]>=2000]
by_year = last_20.groupby("year")["income_per_person"].mean()
by_year.plot(kind = "bar")                                        #using 'kind = "bar"
```

```python
by_year = last_20.groupby("year")["life_expectancy"].mean()
by_year.plot(kind = "barh")                                       #using horizontal bar chart with 'kind = "barh"'
```

## Plotting using Seaborn

### Histogram

```python
sns.histplot (df["income_per_person"], kde = False)    
plt.xlim(0,30000)             # zooming in on the X axis
```

```python
sns.histplot (df["life_expectancy"],kde = False)
```

### Box Plot 

```python
sns.boxplot(df["income_per_person"])
```

```python
sns.boxplot(df["life_expectancy"])
```

### Scatter Plot 

```python
sns.set_style("whitegrid")
sns.scatterplot(x = "income_per_person", y = "life_expectancy", data = df)
```

## Answering questions about the dataset 

### What is the distribution like for income per Person, from last 200 years ? 

```python
sns.histplot(df["income_per_person"], kde = False)
plt.xlim(0, 30000)
plt.xlabel("Income per Person ($, ppp)")
plt.ylabel("Count")
plt.title("Histogram of Income per Person, from last 200 years")
```

### Is there a relationship between fertility rate and ???

```python
sns.scatterplot(
    x = "income_per_person", 
    y = "life_expectancy", 
    hue = "fertility_rate", palette = sns.cubehelix_palette(start = 5,as_cmap = True),
    data = df,
)
```

### What is the highest life expectancy in the data? 

```python
df.sort_values("life_expectancy",ascending = False)
```
